## Task 10

```python
tmdb500=tmdb.sort_values("revenue",ascending=False).head(500)
tmdb500
```

```python
def cen_tend(*s):
    for col in s:
        print(f"central tendencies of {col}: ")
        print(f"mean of {col} : {round(tmdb[col].mean(),2)}")
        print(f"median of {col} : {(tmdb[col].median())}")
        print(f"mode of {col} : {(tmdb[col].mode()[0])}")
        print()
```

```python
cen_tend("budget","revenue","runtime")
```

```python
def box_plot(*s):
    for col in s:
        print("for column :",col)
        sns.boxplot(col,data=tmdb500)
        plt.show()
```

```python
box_plot("budget","revenue","runtime")
```

## 50% of all the movie budgets are lying between
##  0 to 50,000,000 and  more than 120,000,000 are outliers.

## 50% of all the movie revenue are lying between
## 0 to 20,000,000, and more than 40,000,000 are outliers.

## 50% of all the movie runtime are lying between 95 to 130 
## min and more than 155 min and less than 60 min are outliers.
